```python
#Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’.

url= 'http://en.wikipedia.org/wiki/Main_Page'
r=requests.get(url)
soup=bs(r.content)
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

```python
#Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release.
url = 'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc'
r = requests.get(url)
soup=bs(r.text,'html.parser')
movies=soup.findAll('div',{'class':'lister-item mode-advanced'})
movie1=movies[0]
movie1.h3.a.text
movie1.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:5]
movie1.find('div',{'class':'inline-block ratings-imdb-rating'})['data-value']
name=[]
year=[]
rating=[]

for i in movies:
    name.append(i.h3.a.text)
    year.append(i.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:5])
    rating.append(i.find('div',{'class':'inline-block ratings-imdb-rating'})['data-value'])
    
data=list(zip(name,year,rating))
import pandas as pd
df=pd.DataFrame(data,columns=['name','year','rating'])
df.head(50)
```

```python
#Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release).
url='https://www.imdb.com/india/top-rated-indian-movies/'
r=requests.get(url)
soup=bs(r.text,'html.parser')
movies=soup.findAll('td',{'class':'titleColumn'})
movie1=movies[0]
movie1.a.text
movie1.find('span',{'class':'secondaryInfo'}).text[1:5]
movie1.find('td',{'class':'ratingColumn imdbRating'})
name=[]
year=[]
rating=[]

for i in movies:
    name.append(i.a.text)
    year.append(i.find('span',{'class':'secondaryInfo'}).text[1:5])
    rating.append(i.find('td',{'class':'ratingColumn imdbRating'}))
    
data=list(zip(name,year,rating))
df=pd.DataFrame(data,columns=['name','year','rating'])
df.head(100)
```

```python
#Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’
url="https://bookpage.com/search?book_genre=fiction&contributor_id=953&page=1&type=Review"
r=requests.get(url)
soup=bs(r.text,'lxml')
book=soup.findAll('div',{'class':'bp-block article-info'})   #title
book1=book[0]
book1.h4.a.text
book1.find('div',{'class':'flex-article-content'}).p.text[1:14]  #author
book1.find('p',{'class','genre-links hidden-phone'}).a.text  #genre
book1.find('p',{'class':'excerpt'}).text  #review
title=[]
author=[]
genre=[]
review=[]

for i in book:
    title.append(i.h4.a.text)
    author.append(i.find('div',{'class':'flex-article-content'}).p.text[1:14])
    genre.append(i.find('p',{'class','genre-links hidden-phone'}).a.text)
    review.append(book1.find('p',{'class':'excerpt'}).text.replace('\n',''))
    
data=list(zip(title,author,genre,review))
df=pd.DataFrame(data,columns=['title','author','genre','review'])
df.head()
```

```python
#ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
url='https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
r=requests.get(url).text
soup=bs(r,'lxml')
table=soup.find('div',attrs={'class':'rankings-block__container'})
playerlist=[]
for tr in table('tr'):
    for th in table('th'):
        for td in table('td'):
            playerlist.append(td.text.replace('\n','').strip())
            
for player in playerlist:
    print(player)
```

```python
#Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.
url='https://www.amazon.in/s?k=mobile+phones&i=electronics&rh=n%3A1389401031%2Cp_36%3A-2000000&s=price-asc-rank&qid=1632904480&rnid=1318502031&ref=sr_nr_p_36_5'
r=requests.get(url).text
soup=bs(r,'lxml')
mobile=soup.findAll('h2',{'class':'a-size-mini a-spacing-none a-color-base s-line-clamp-2'})

phone=[]
for i in mobile:
    phone.append(i.text)
phone

rate=soup.findAll('span',{'class':'a-price-whole'})

price=[]
for i in rate:
    price.append(i.text)
price

review=soup.findAll('div',{'class':'a-row a-size-small'})

rating=[]
for i in review:
    rating.append(i.text)
rating

links=soup.select('div',{'class':'a-section aok-relative s-image-fixed-height'})
images=[x.find('img') for x in links]
images

amazon=pd.DataFrame({})
amazon['phone']=phone
amazon['price']=price
amazon['rating']=rating
amazon
```

```python
#Write a python program to extract information about the local weather from the National Weather Service website of USA,
#https://www.weather.gov/ for the city, San Francisco.
#You need to extract data about 7 day extended forecast display for the city.
#The data should include period, short description, temperature and description.

url='https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YVQ_57gzZPY'
r=requests.get(url).text
soup=bs(r,'html.parser')
#period
a=soup.findAll('div',{'class':'col-sm-2 forecast-label'})
period=[]
for i in a:
    period.append(i.text)
period
short_desc=[]
for i in soup.find_all('div',class_="col-sm-10 forecast-text"):        
        short_desc.append(i.text.split(',')[0])


#temperature and description
c=soup.findAll('div',{'class':'col-sm-10 forecast-text'})
temp_disc=[]
for i in c:
    temp_disc.append(i.text)
temp_disc

weather=pd.DataFrame({})
weather['period']=period
weather['temp_disc']=temp_disc
weather['short_desc']=short_desc
weather 
```

```python
#Write a python program to scrape fresher job listings from ‘https://internshala.com/’. 
#It should include job title, company name, CTC, and apply date

url='https://internshala.com/internships/work-from-home-jobs'
r=requests.get(url).text
soup=bs(r,'lxml')
#title
a=soup.findAll('div',{'class':'heading_4_5 profile'})
job_title=[]
for i in a:
    job_title.append(i.text.replace('\n',''))
job_title
#company_name
b=soup.findAll('div',{'class':'heading_6 company_name'})
company_name=[]
for i in b:
    company_name.append(i.text[25:65].replace('\n',''))
company_name
#ctc
c=soup.findAll('span',class_='stipend')
c
package=[]
for i in c:
    package.append(i.text)
package
#apply_date
date=soup.find('div',{'class':'apply_by'}).find('div',{'class':'item_body'})
date
apply=[]
for i in date:
    apply.append(i)
apply
df=pd.DataFrame({})
df['job_title']=job_title
df['company_name']=company_name
df['package']=package
df
```
